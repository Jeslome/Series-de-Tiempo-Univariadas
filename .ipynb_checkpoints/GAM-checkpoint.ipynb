# Modelos GAM para la serie de Atleta Iv√°n

```python
import pandas as pd
from pygam import GAM, s, f
import statsmodels.api as sm
import numpy as np
from matplotlib import pyplot
from sklearn.metrics import mean_squared_error
```

### Cargar datos

```python
ruta = 'Ivan - Serie de Tiempo.xlsx'

# Lee el archivo Excel
ivan = pd.read_excel(ruta,usecols="A,B")

# Muestra las primeras filas del DataFrame
print(ivan.tail())
```

```python
indice = pd.date_range(start='31/12/2018', periods=268, freq='W-MON')
ivants =pd.Series(ivan['KMS'].values,index=indice)

ivants.plot(figsize=(8,5))
pyplot.show()
```

<h1 style="color: purple;">Modelo m√°s simple que podemos usar en GAM</h1>

```python
X_ivan = ivan[['DATE']].values  # X con las columnas 'DATE'
y_ivan = ivan['KMS'].values  # Variable dependiente 

# Ajustar el modelo GAM con suavizados para 'DATE'
gam = GAM(s(0))  # 's(0)' corresponde a 'DATE' ..... gam(y_ivan~x_ivan) familia normal con link identidad
gam.fit(X_ivan, y_ivan)

GAM.summary(gam)
```

```python
X_pred = X_ivan
y_pred = gam.predict(X_pred)

pyplot.plot(X_ivan, y_ivan, label="Datos observados")
pyplot.plot(X_pred, y_pred, color='purple', label="Predicci√≥n GAM")
pyplot.xlabel("week")
pyplot.ylabel("KMS")
pyplot.legend()
pyplot.show()
```

EL modelo subestima los picos de todas las semanas. Si hacemos previsiones con este modelo, este fenomeno de sibestimaci√≥n continuar√°, pero la serie real aumentar√°. Por lo tanto, este modelo no es muy bueno y nuestro objetivo es mejorarlo

### Calculamos el RMSE para este primer modelo

```python
# Calcular el RMSE
rmse = mean_squared_error(y_ivan, y_pred, squared=False)

print(f'El error cuadr√°tico medio (RMSE) es: {rmse}')
```

<h1 style="color: purple;">Modelos GAM y sus variaciones</h1>

basis="ps" son bases p-spline  y  basis="cp" son bases p-spline c√≠clica, √∫til para construir funciones peri√≥dicas.

```python
gam1 = GAM(s(0, basis='cp', n_splines=52)) # base de splines ciclicos penalizados.
                                           # Descripci√≥n: Modela cambios abruptos y discontinuidades
                                           # Para 52 semanas
gam2 = GAM(s(0, basis='cp', n_splines=17)) # base de splines c√∫bicos penalizados.
                                           # Para 17 semanas

gam3 = GAM(s(0, n_splines=52)) # T√©rmino suavizado peri√≥dico.
                               # Descripci√≥n: Modela datos c√≠clicos, como semanas o meses, repitiendo el patr√≥n cada ciclo.
                               # Por eso se usa 52, que seria las semanas.
gam4 = GAM(s(0, n_splines=17)) # Se encontro que el periodo era casi 17 semanas
```

```python
gam1.fit(X_ivan, y_ivan)
GAM.summary(gam1)
```

```python
X_pred1 = X_ivan
y_pred1 = gam1.predict(X_pred1)

pyplot.plot(X_ivan, y_ivan, label="Datos observados")
pyplot.plot(X_pred1, y_pred1, color='purple', label="Predicci√≥n GAM")
pyplot.xlabel("week")
pyplot.ylabel("KMS")
pyplot.legend()
pyplot.show()
```

```python
gam2.fit(X_ivan, y_ivan)
GAM.summary(gam2)
```

```python
X_pred2 = X_ivan
y_pred2 = gam2.predict(X_pred2)

pyplot.plot(X_ivan, y_ivan, label="Datos observados")
pyplot.plot(X_pred2, y_pred2, color='purple', label="Predicci√≥n GAM")
pyplot.xlabel("week")
pyplot.ylabel("KMS")
pyplot.legend()
pyplot.show()
```

```python
gam3.fit(X_ivan, y_ivan)
GAM.summary(gam3)
```

```python
X_pred3 = X_ivan
y_pred3 = gam3.predict(X_pred3)

pyplot.plot(X_ivan, y_ivan, label="Datos observados")
pyplot.plot(X_pred3, y_pred3, color='purple', label="Predicci√≥n GAM")
pyplot.xlabel("week")
pyplot.ylabel("KMS")
pyplot.legend()
pyplot.show()
```

```python
gam4.fit(X_ivan, y_ivan)
GAM.summary(gam4)
```

```python
X_pred4 = X_ivan
y_pred4 = gam4.predict(X_pred4)

pyplot.plot(X_ivan, y_ivan, label="Datos observados")
pyplot.plot(X_pred4, y_pred4, color='purple', label="Predicci√≥n GAM")
pyplot.xlabel("week")
pyplot.ylabel("KMS")
pyplot.legend()
pyplot.show()
```

### Calculamos el RMSE para cada modelo

```python
# Calcular el RMSE
rmse1 = mean_squared_error(y_ivan, y_pred1, squared=False)
rmse2 = mean_squared_error(y_ivan, y_pred2, squared=False)
rmse3 = mean_squared_error(y_ivan, y_pred3, squared=False)
rmse4 = mean_squared_error(y_ivan, y_pred4, squared=False)

print(f'El error cuadr√°tico medio (RMSE) para el modelo gam1 es: {rmse1}')
print(f'El error cuadr√°tico medio (RMSE) para el modelo gam2 es: {rmse2}')
print(f'El error cuadr√°tico medio (RMSE) para el modelo gam3 es: {rmse3}')
print(f'El error cuadr√°tico medio (RMSE) para el modelo gam4 es: {rmse4}')
```

<h1 style="color: purple;">Modelos GAM con ARIMA</h1>

<h3 style="color: purple;">Agregamos un AR(17) al mejor modelo GAM </h3>

Vemos que el mejor ha sido gam1, un modelo base de splines c√∫bicos penalizados, con 52 splines 

Los tres par√°metros en ARIMA(order=(p, d, q)) son:<br>
 p: El n√∫mero de t√©rminos autorregresivos (AR).<br>
 d: El grado de diferenciaci√≥n (I para integrado). Si d=0, el modelo no realiza diferenciaci√≥n, lo que es el caso de los modelos ARMA.<br>
 q: El n√∫mero de t√©rminos de promedio m√≥vil (MA).

```python
# Predecir usando gam1
y_pred_gam = gam1.predict(X_ivan)

# Calcular los residuales
residuals = y_ivan - y_pred_gam

# Ajustar AR(17) para los residuales
arima_model = sm.tsa.ARIMA(residuals, order=(17, 0, 0))
arima_fit = arima_model.fit()

# Predecir residaules AR(17) 
arima_predictions = arima_fit.predict()

final_predictions = y_pred_gam + arima_predictions
```

```python
pyplot.plot(X_ivan, y_ivan, label="Datos observados")

pyplot.plot(X_ivan, y_pred_gam, label="Predicci√≥n GAM", color='purple')

pyplot.plot(X_ivan, final_predictions, label="Predicci√≥n GAM + AR(17)", color='red')

pyplot.xlabel("week")
pyplot.ylabel("KMS")
pyplot.legend()
pyplot.show()
```

```python
rmse5 = mean_squared_error(y_ivan, final_predictions, squared=False)
print(f'El error cuadr√°tico medio (RMSE) para el modelo gam5 es: {rmse5}')
```

## Agregamos un AR(52) al mejor modelo GAM 

```python
# Predecir usando gam1
y_pred_gam1 = gam1.predict(X_ivan)

# Calcular los residuales
residuals1 = y_ivan - y_pred_gam1

# Ajustar AR(52) para los residuales
arima_model1 = sm.tsa.ARIMA(residuals1, order=(52, 0, 0)) 
arima_fit1 = arima_model1.fit()

# Predecir residaules AR(17) 
arima_predictions1 = arima_fit1.predict()

final_predictions1 = y_pred_gam1 + arima_predictions1

pyplot.plot(X_ivan, y_ivan, label="Datos observados")

pyplot.plot(X_ivan, y_pred_gam1, label="Predicci√≥n GAM", color='purple')

pyplot.plot(X_ivan, final_predictions1, label="Predicci√≥n GAM + AR(52)", color='red')

pyplot.xlabel("week")
pyplot.ylabel("KMS")
pyplot.legend()
pyplot.show()
```

```python
rmse6 = mean_squared_error(y_ivan, final_predictions1, squared=False)
print(f'El error cuadr√°tico medio (RMSE) para el modelo gam5 es: {rmse6}')
```

## Agregamos un ARMA(52,1) al mejor modelo GAM 

```python
# Predecir usando gam1
y_pred_gam2 = gam1.predict(X_ivan)

# Calcular los residuales
residuals2 = y_ivan - y_pred_gam2

# Ajustar AR(52) para los residuales
arima_model2 = sm.tsa.ARIMA(residuals1, order=(52, 0, 1)) 
arima_fit2 = arima_model2.fit()

# Predecir residaules AR(17) 
arima_predictions2 = arima_fit2.predict()

final_predictions2 = y_pred_gam2 + arima_predictions2

pyplot.plot(X_ivan, y_ivan, label="Datos observados")

pyplot.plot(X_ivan, y_pred_gam1, label="Predicci√≥n GAM", color='purple')

pyplot.plot(X_ivan, final_predictions2, label="Predicci√≥n GAM + ARIMA(52,0,1)", color='red')

pyplot.xlabel("week")
pyplot.ylabel("KMS")
pyplot.legend()
pyplot.show()
```

```python
rmse7 = mean_squared_error(y_ivan, final_predictions2, squared=False)
print(f'El error cuadr√°tico medio (RMSE) para el modelo gam5 es: {rmse7}')
```

## Agregamos un ARMA(52,2) al mejor modelo GAM 

```python
# Predecir usando gam1
y_pred_gam3 = gam1.predict(X_ivan)

# Calcular los residuales
residuals3 = y_ivan - y_pred_gam3

# Ajustar AR(52) para los residuales
arima_model3 = sm.tsa.ARIMA(residuals3, order=(52, 0, 2)) 
arima_fit3 = arima_model3.fit()

# Predecir residaules AR(17) 
arima_predictions3 = arima_fit3.predict()

final_predictions3 = y_pred_gam3 + arima_predictions3

pyplot.plot(X_ivan, y_ivan, label="Datos observados")

pyplot.plot(X_ivan, y_pred_gam1, label="Predicci√≥n GAM", color='purple')

pyplot.plot(X_ivan, final_predictions3, label="Predicci√≥n GAM + ARIMA(52,0,2)", color='red')

pyplot.xlabel("week")
pyplot.ylabel("KMS")
pyplot.legend()
pyplot.show()
```

```python
rmse8 = mean_squared_error(y_ivan, final_predictions3, squared=False)
print(f'El error cuadr√°tico medio (RMSE) para el modelo gam5 es: {rmse8}')
```
