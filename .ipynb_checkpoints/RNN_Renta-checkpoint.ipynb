<h3 style="color: #e67e22; text-align: left;">Creación de retardos</h3>

````python
def get_lagged_features(df, n_steps, n_steps_ahead):
    """
    df: pandas DataFrame of time series to be lagged
    n_steps: number of lags, i.e. sequence length
    n_steps_ahead: forecasting horizon
    """
    lag_list = []
    n_steps=12
    
    for lag in range(n_steps + n_steps_ahead - 1, n_steps_ahead - 1, -1):
        lag_list.append(df.shift(lag))
    lag_array = np.dstack([i[n_steps+n_steps_ahead-1:] for i in lag_list])
    lag_array = np.swapaxes(lag_array, 1, -1)
    return lag_array
````

````python
n_steps_ahead
````

###### Crear las características (features)
1.) La función get_lagged_features se utiliza para crear un conjunto de características utilizando los datos anteriores a cada punto en la serie temporal.<br>
2.) x_train y x_test contendrán las características retardadas (lags) de los datos de entrenamiento y prueba. Básicamente, esto significa que para cada punto de tiempo en la serie temporal, se toman n_steps puntos anteriores como entrada.<br><br>

###### Obtener las etiquetas (targets)
1.) Aquí, y_train y y_test son los valores que se desean predecir.<br>
2.) La expresión n_steps + n_steps_ahead - 1 se utiliza para calcular el índice a partir del cual se extraen los valores objetivo. Este índice asegura que las etiquetas correspondan a los pasos de tiempo correctos después de los retardos y la ventana de predicción.<br><br>

###### Obtener las marcas de tiempo (timestamps)
1.) y_train_timestamps y y_test_timestamps contienen los índices de tiempo asociados a cada valor en y_train y y_test<br>
2.) Esto es útil si se desea comparar las predicciones con los valores reales en una serie temporal.

````python
x_train = get_lagged_features(df_train, n_steps, n_steps_ahead)
y_train =  df_train.values[n_steps + n_steps_ahead - 1:]
y_train_timestamps = df_train.index[n_steps + n_steps_ahead - 1:]

x_test = get_lagged_features(df_test, n_steps, n_steps_ahead)
y_test =  df_test.values[n_steps + n_steps_ahead - 1:]
y_test_timestamps = df_test.index[n_steps + n_steps_ahead - 1:]
````

<h3 style="color: #e67e22; text-align: left;">Especificación del modelo</h3>

Se define tres funciones diferentes que crean y devuelven modelos secuenciales de redes neuronales recurrentes (RNN) en TensorFlow utilizando la API Keras. Cada función construye un tipo específico de RNN: SimpleRNN, GRU, y LSTM, cada uno hace lo siguiente:<br><br>

#### `SimpleRNN_`
Crea un modelo secuencial con una capa de `SimpleRNN` y una capa densa (`Dense`).

- **Parámetros:**
  - **`n_units`**: Número de neuronas en la capa `SimpleRNN`.
  - **`l1_reg`**: Coeficiente de regularización L1 para evitar el sobreajuste.
  - **`seed`**: Semilla para inicializar pesos de manera reproducible.
  
- **Capa RNN**: Utiliza `SimpleRNN`, con una activación `tanh` y diversas inicializaciones de pesos y bias controladas por la semilla.
- **Capa de salida**: Capa densa con un único neurón (para regresión) y una regularización L1.
- **Compilación**: Utiliza la función de pérdida `mean_squared_error` y el optimizador `adam`.

#### `GRU_`
Similar al anterior, pero en lugar de `SimpleRNN` usa `GRU`.

- **Capa RNN**: Utiliza `GRU`, que es más avanzada y eficaz que `SimpleRNN` para capturar dependencias a largo plazo en series temporales.

#### `LSTM_`
En lugar de `SimpleRNN` o `GRU`, utiliza `LSTM`.

- **Capa RNN**: Utiliza `LSTM`, que es ideal para capturar dependencias a largo plazo en secuencias, como series temporales complejas.


````python
def SimpleRNN_(n_units = 10, l1_reg=0, seed=0):
  model = keras.models.Sequential()
  model.add(keras.layers.SimpleRNN(n_units, activation='tanh', kernel_initializer=keras.initializers.glorot_uniform(seed), bias_initializer=keras.initializers.glorot_uniform(seed), recurrent_initializer=keras.initializers.orthogonal(seed), kernel_regularizer=keras.regularizers.l1(l1_reg), input_shape=(x_train.shape[1], x_train.shape[-1]), unroll=True, stateful=False))  
  model.add(keras.layers.Dense(1, kernel_initializer=keras.initializers.glorot_uniform(seed), bias_initializer=keras.initializers.glorot_uniform(seed), kernel_regularizer=keras.regularizers.l1(l1_reg)))
  model.compile(loss='mean_squared_error', optimizer='adam')
  return model


def GRU_(n_units = 10, l1_reg=0, seed=0):
  model = keras.models.Sequential()
  model.add(keras.layers.GRU(n_units, activation='tanh', kernel_initializer=keras.initializers.glorot_uniform(seed), bias_initializer=keras.initializers.glorot_uniform(seed), recurrent_initializer=keras.initializers.orthogonal(seed), kernel_regularizer=keras.regularizers.l1(l1_reg), input_shape=(x_train.shape[1], x_train.shape[-1]), unroll=True))  
  model.add(keras.layers.Dense(1, kernel_initializer=keras.initializers.glorot_uniform(seed), bias_initializer=keras.initializers.glorot_uniform(seed), kernel_regularizer=keras.regularizers.l1(l1_reg)))
  model.compile(loss='mean_squared_error', optimizer='adam')
  return model


def LSTM_(n_units = 10, l1_reg=0, seed=0):
  model = keras.models.Sequential()
  model.add(keras.layers.LSTM(n_units, activation='tanh', kernel_initializer=keras.initializers.glorot_uniform(seed), bias_initializer=keras.initializers.glorot_uniform(seed), recurrent_initializer=keras.initializers.orthogonal(seed), kernel_regularizer=keras.regularizers.l1(l1_reg), input_shape=(x_train.shape[1], x_train.shape[-1]), unroll=True)) 
  model.add(keras.layers.Dense(1, kernel_initializer=keras.initializers.glorot_uniform(seed), bias_initializer=keras.initializers.glorot_uniform(seed), kernel_regularizer=keras.regularizers.l1(l1_reg)))
  model.compile(loss='mean_squared_error', optimizer='adam')
  return model
````

Usaremos un tamaño del lote igual a 7 y 50 epochs tal como se hizo en las redes neuronales multicapa.<br>
En resumen, el modelo se entrenará en 50 ciclos completos a través del conjunto de datos, y en cada ciclo, los datos se procesarán en grupos de 7 muestras a la vez.

````python
max_epochs = 50 
batch_size = 7
````

Este callback se utiliza para evitar el sobreentrenamiento (overfitting) al monitorear la pérdida y detener el entrenamiento si no hay mejoras significativas durante un número determinado de épocas (patience). También asegura que los pesos del modelo sean los mejores obtenidos durante el entrenamiento, restaurándolos al estado en el que la pérdida era mínima.

````python
es = keras.callbacks.EarlyStopping(monitor='loss', mode='min', verbose=1, patience=10, min_delta=1e-7, restore_best_weights=True)
````

Este diccionario organiza y configura los parámetros para los modelos de SimpleRNN, GRU, y LSTM, permitiendo una fácil referencia y modificación de estos parámetros al entrenar y evaluar diferentes tipos de modelos de redes neuronales recurrentes

<h3 style="color: #e67e22; text-align: left;">Validación cruzada</h3>

### Explicación

#### `cross_val = True:`
Establece la variable `cross_val` en `True`. Esto indica que se realizará una validación cruzada durante el entrenamiento. La advertencia señala que cambiar esto a `True` hará que el proceso sea muy largo.

#### Condición `if do_training and cross_val:`
Verifica si `do_training` y `cross_val` están en `True`. Si ambos son `True`, el bloque de código dentro de esta condición se ejecutará. Si `do_training` es `False`, no se ejecutará.

#### Configuración de parámetros:
- **`n_units = [5, 10, 20]:`** Lista de posibles valores para el número de unidades en las capas recurrentes.
- **`l1_reg = [0.0, 0.001, 0.01, 0.1]:`** Lista de posibles valores para la regularización L1.
- **`param_grid = {'n_units': n_units, 'l1_reg': l1_reg}:`** Crea un diccionario con los parámetros y sus valores posibles para la búsqueda en la cuadrícula (`GridSearch`).

#### `tscv = TimeSeriesSplit(n_splits = 5):`
Configura una validación cruzada específica para series temporales con 5 particiones (`splits`). `TimeSeriesSplit` divide los datos en `n_splits` partes, usando los primeros `k` pliegues para entrenamiento y el `(k+1)`-ésimo pliegue para prueba.

#### Búsqueda en la cuadrícula para cada modelo:
- **`for key in params.keys()::`** Itera sobre cada clave en el diccionario `params` (para cada tipo de modelo).
- **`print('Performing cross-validation. Model:', key):`** Imprime un mensaje indicando que se está realizando la validación cruzada para el modelo actual.
- **`model = keras.wrappers.scikit_learn.KerasRegressor(...):`** Crea un `KerasRegressor` usando la función del modelo correspondiente, con el número máximo de épocas y tamaño de lote especificado.
- **`grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=tscv, n_jobs=1, verbose=2):`** Configura una búsqueda en la cuadrícula para encontrar la mejor combinación de parámetros utilizando validación cruzada en series temporales.
- **`grid_result = grid.fit(x_train, y_train, callbacks=[es]):`** Ajusta el modelo a los datos de entrenamiento y realiza la búsqueda en la cuadrícula. Utiliza el callback `EarlyStopping`.
- **`print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_)):`** Imprime la mejor puntuación y los parámetros que produjeron esa puntuación.
- **`means = grid_result.cv_results_['mean_test_score']:`** Obtiene las puntuaciones medias de prueba para cada conjunto de parámetros.
- **`stds = grid_result.cv_results_['std_test_score']:`** Obtiene las desviaciones estándar de las puntuaciones de prueba.
- **`params_ = grid_result.cv_results_['params']:`** Obtiene los parámetros correspondientes a cada puntuación.
- **`for mean, stdev, param_ in zip(means, stds, params_)::`** Imprime la puntuación media, desviación estándar y parámetros para cada combinación.
- **`params[key]['H'] = grid_result.best_params_['n_units']:`** Guarda el mejor valor de `n_units` para el modelo en el diccionario `params`.
- **`params[key]['l1_reg'] = grid_result.best_params_['l1_reg']:`** Guarda el mejor valor de `l1_reg` para el modelo en el diccionario `params`.


````python
cross_val = True # WARNING: Changing this to True will take many hours to run

if do_training and cross_val:
    n_units = [5, 10, 20]
    l1_reg = [0.0, 0.001, 0.01, 0.1]
    
    # A dictionary containing a list of values to be iterated through
    # for each parameter of the model included in the search
    param_grid = {'n_units': n_units, 'l1_reg': l1_reg}
    
    # In the kth split, TimeSeriesSplit returns first k folds 
    # as training set and the (k+1)th fold as test set.
    tscv = TimeSeriesSplit(n_splits = 5)
    
    # A grid search is performed for each of the models, and the parameter set which
    # performs best over all the cross-validation splits is saved in the `params` dictionary
    for key in params.keys():
        print('Performing cross-validation. Model:', key)
        model = keras.wrappers.scikit_learn.KerasRegressor(build_fn=params[key]['function'], epochs=max_epochs, 
                               batch_size=batch_size, verbose=2)
        #from scikeras.wrappers import KerasClassifier, KerasRegressor
        grid = GridSearchCV(estimator=model, param_grid=param_grid, 
                            cv=tscv, n_jobs=1, verbose=2)
        grid_result = grid.fit(x_train, y_train, callbacks=[es])
        print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
        
        means = grid_result.cv_results_['mean_test_score']
        stds = grid_result.cv_results_['std_test_score']
        params_ = grid_result.cv_results_['params']
        for mean, stdev, param_ in zip(means, stds, params_):
            print("%f (%f) with %r" % (mean, stdev, param_))
            
        params[key]['H'] = grid_result.best_params_['n_units']
        params[key]['l1_reg']= grid_result.best_params_['l1_reg']
````

<h2 style="color: #e67e22; text-align: left;">Ajustar el modelo con los mejores parámetros</h2>

#### `for key in params.keys():`
- Itera sobre cada clave en el diccionario `params`, que representa cada tipo de modelo (`RNN`, `GRU`, `LSTM`).

#### `tf.random.set_seed(0):`
- Establece una semilla para la generación de números aleatorios en TensorFlow. Esto asegura que los resultados sean reproducibles.

#### `print('Training', key, 'model'):`
- Imprime un mensaje indicando que se está entrenando el modelo correspondiente a la clave actual.

#### `model = params[key]['function'](params[key]['H'], params[key]['l1_reg']):`
- Llama a la función del modelo asociada con la clave actual (`SimpleRNN_`, `GRU_`, `LSTM_`), pasando el número de unidades (`params[key]['H']`) y la regularización L1 (`params[key]['l1_reg']`) como argumentos. Crea una instancia del modelo.

#### `model.fit(x_train, y_train, epochs=max_epochs, batch_size=batch_size, callbacks=[es], shuffle=False):`
- Ajusta el modelo a los datos de entrenamiento (`x_train`, `y_train`). Se entrena el modelo durante `max_epochs` épocas con un tamaño de lote (`batch_size`) especificado. Utiliza el callback `EarlyStopping` (`es`) para detener el entrenamiento si el modelo deja de mejorar. `shuffle=False` asegura que los datos no se mezclen antes de cada época, lo cual es importante para series temporales.

#### `params[key]['model'] = model:`
- Guarda el modelo entrenado en el diccionario `params` bajo la clave correspondiente, para su uso posterior o evaluación.


````python
if do_training is True:
    for key in params.keys():
        tf.random.set_seed(0)
        print('Training', key, 'model')
        model = params[key]['function'](params[key]['H'], params[key]['l1_reg'])
        model.fit(x_train, y_train, epochs=max_epochs, 
                  batch_size=batch_size, callbacks=[es], shuffle=False)
        params[key]['model'] = model
````

- Itera sobre cada clave en el diccionario `params`, que representa cada tipo de modelo (`RNN`, `GRU`, `LSTM`).

#### `model = params[key]['model']:`
- Recupera el modelo previamente entrenado almacenado en el diccionario `params` bajo la clave actual.

#### `model.summary():`
- Imprime un resumen del modelo, que incluye la arquitectura de las capas, el número de parámetros, y otras características del modelo.

#### `params[key]['pred_train'] = model.predict(x_train, verbose=1):`
- Realiza predicciones sobre los datos de entrenamiento (`x_train`) usando el modelo. El resultado se almacena en el diccionario `params` bajo la clave actual como `pred_train`. `verbose=1` muestra información sobre el progreso de la predicción.

#### `params[key]['MSE_train'] = mean_squared_error(y_train, params[key]['pred_train']):`
- Calcula el error cuadrático medio (MSE) entre las predicciones del modelo (`pred_train`) y las verdaderas etiquetas de entrenamiento (`y_train`). Este valor se almacena en el diccionario `params` bajo la clave actual como `MSE_train`.

#### `params[key]['pred_test'] = model.predict(x_test, verbose=1):`
- Realiza predicciones sobre los datos de prueba (`x_test`) usando el modelo. El resultado se almacena en el diccionario `params` bajo la clave actual como `pred_test`. `verbose=1` muestra información sobre el progreso de la predicción.

#### `params[key]['MSE_test'] = mean_squared_error(y_test, params[key]['pred_test']):`
- Calcula el error cuadrático medio (MSE) entre las predicciones del modelo (`pred_test`) y las verdaderas etiquetas de prueba (`y_test`). Este valor se almacena en el diccionario `params` bajo la clave actual como `MSE_test`.


````python
for key in params.keys():
    model = params[key]['model']
    model.summary()
    
    params[key]['pred_train'] = model.predict(x_train, verbose=1)
    params[key]['MSE_train'] = mean_squared_error(y_train, params[key]['pred_train'])
    
    params[key]['pred_test'] = model.predict(x_test, verbose=1) 
    params[key]['MSE_test'] = mean_squared_error(y_test, params[key]['pred_test'])
````
