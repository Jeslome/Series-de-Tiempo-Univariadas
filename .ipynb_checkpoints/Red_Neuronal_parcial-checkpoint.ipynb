Aquí rechazamos la hipótesis nula ya que el valor p es menor que 0.01, por lo tanto,  podemos rechazar la prueba al nivel de confianza del 99%. Esto sugiere que la serie temporal  es **estacionaria**.

**Detección de datos atípicos para crear variables de intervención**

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
```

```python
# Ajustar un modelo ARIMA
model = sm.tsa.ARIMA(df_box['diferenciado'], order=(1, 0, 1))
fit = model.fit()

# Obtener residuos
residuals = fit.resid

# Calcular t-stat
t_stat = residuals / np.std(residuals)

# Identificar outliers basados en un umbral
threshold = 3  # Umbral para detectar outliers
outliers = np.where(np.abs(t_stat) > threshold)[0]

# Mostrar resultados
print("Índices de outliers encontrados:", outliers)
print("Fechas de outliers encontrados:")
print(df.index[outliers])

# Graficar resultados
plt.figure(figsize=(12, 6))
plt.plot(df['valor'], label='Serie Original')
plt.scatter(df.index[outliers], df['valor'].iloc[outliers], color='red', label='Outliers', zorder=5)
plt.title('Detección de Outliers en la Serie de Tiempo')
plt.legend()
plt.show()
```

### Incorporando variables necesarias

**Detección de  outliers**

```python
# Ajustar un modelo ARIMA
model = sm.tsa.ARIMA(df_box['trabajadores_box'], order=(1, 1, 1))
fit = model.fit()

# Obtener residuos
residuals = fit.resid

# Calcular t-stat
t_stat = residuals / np.std(residuals)

# Identificar outliers basados en un umbral
threshold = 3  # Umbral para detectar outliers
outliers = np.where(np.abs(t_stat) > threshold)[0]

# Mostrar resultados
print("Índices de outliers encontrados:", outliers)
print("Fechas de outliers encontrados:")
print(df_box.index[outliers])

# Graficar resultados
plt.figure(figsize=(12, 6))
plt.plot(df_box['trabajadores_box'], label='Serie Original')
plt.scatter(df_box.index[outliers], df_box['trabajadores_box'].iloc[outliers], color='red', label='Outliers', zorder=5)
plt.title('Detección de Outliers en la Serie de Tiempo')
plt.legend()
plt.show()
```

**Incorporando variables de intervención de acuerdo al outlier detectado**

```python
df_box['intervencion'] = 0  # Inicializar en 0
df_box['intervencion'].iloc[outliers[1]] = 1  
```

```python
df_box
```

**Creando las variables rezagadas**

```python
for lag in [1, 3, 6, 12]:
    df_box[f't-{lag}'] = df_box['diferenciado_1'].shift(lag)

# Mostrar el resultado
print(df_box)
```

Incorporando las componentes de fourioer para la estacionalidad anual.

```python
df_box['fecha']=df_box.index
date_time=df_box['fecha']
type(date_time)
date_time.map(pd.Timestamp.timestamp) 
timestamp_s = date_time.map(pd.Timestamp.timestamp)
timestamp_s
```

```python
# Periodos
anio = 365.25 * 24 * 60 * 60  # segundos en un año

# Componentes estacionales anuales
df_box['Year sin'] = np.sin(timestamp_s * (2 * np.pi / anio))
df_box['Year cos'] = np.cos(timestamp_s * (2 * np.pi / anio))
```

```python
plt.plot(np.array(df_box['Year sin'])[:100])
plt.plot(np.array(df_box['Year cos'])[:100])
plt.xlabel('Time [h]')
plt.title('Time of year signal')
```

```python
df_box
```

```python
# Agregar componentes de Fourier
for i in range(1, 4):  # Incorporar 3 componentes
    df_box[f'Year sin {i}'] = np.sin(timestamp_s * (2 * np.pi * i / anio))
    df_box[f'Year cos {i}'] = np.cos(timestamp_s * (2 * np.pi * i / anio))

# Mostrar el DataFrame
print(df_box.head())

# Graficar algunas componentes
plt.figure(figsize=(12, 6))
plt.plot(df_box.index, df_box['Year sin 1'], label='Year sin 1')
plt.plot(df_box.index, df_box['Year cos 1'], label='Year cos 1')
plt.plot(df_box.index, df_box['Year sin 2'], label='Year sin 2')
plt.plot(df_box.index, df_box['Year cos 2'], label='Year cos 2')
plt.title('Componentes de Fourier')
plt.legend()
plt.show()
```
